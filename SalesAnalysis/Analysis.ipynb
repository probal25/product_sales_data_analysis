# Cleaning data set

##### Drop rows of nan

```python
nan_df = df[df.isna().any(axis=1)]
nan_df.shape
```

```python
df = df.dropna(how='all')
df.head()
```

##### Find 'Or' and delete it

```python
df = df[df['Order Date'].str[0:2] != 'Or']
```

##### Convert coulmns to the right type

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

## Augment data with additional columns

#### Add Month column

```python
df['Month'] = df['Order Date'].str[0:2]
df['Month'] = df['Month'].astype('int32')
df.head()
```

#### Add a Sales cloumn

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

##### Add a City column

```python
# Get city name

def get_city_name(address):
    return address.split(',')[1]

def get_state_name(address):
    return address.split(',')[2].split(' ')[1]
# apply() function  

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city_name(x)}({get_state_name(x)})")
df.head()
```

### Task 1 : What's the best month for sales? How much was earned that month?

```python
result = df.groupby('Month').sum()
result
```

##### Ploting

```python
months = range(1, 13)

plt.bar(months, result['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD($)')
plt.xlabel('Month number')
plt.show()
```

#### Task 3: What time should we  display advertisement to maximize likelihood of customer's buying product?

```python
# Make Order date column to DateTime object

df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
# Adding Hour column and Minute column into the data frame

df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute

df.head()
```

##### plotting

```python
hours = [hour for hour, i in df.groupby('Hour')]

plt.plot(hours, df.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel("Hours")
plt.ylabel("Number of Orders")
plt.grid()
plt.show()
```

#### Task 4: What products are most often sold togather 

```python
f_df = df[df['Order ID'].duplicated(keep=False)]
f_df['Grouped'] = f_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

f_df = f_df[['Order ID', 'Grouped']].drop_duplicates()

f_df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in f_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(20):
    print(key, value)
    
    
```

#### Task 5: What product sold the most? What's the reason for that?

```python
df.head()
```

```python
product_group = df.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']


products = [product for product, i in product_group]

plt.bar(products, quantity_ordered)

plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.xticks(products, rotation = 'vertical', size=8)

plt.show()
```

```python
prices = df.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1. twinx()

ax1.bar(products, quantity_ordered, color = 'g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color = 'g')
ax2.set_ylabel('Prices ($)', color = 'b')
ax1.set_xticklabels(products, rotation='vertical', size = 8)


plt.show()
```
